```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
type(iris)
```

```python
data = iris.data
```

```python
data.shape
```

```python
iris.feature_names
```

```python
targets = iris.target
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 3)
```

```python
new_data = pca.fit_transform(data)
```

```python
new_data.shape
```

```python
import pandas as pd
```

```python
new_data = pd.DataFrame(new_data)
```

```python
new_data.head()
```

```python
new_data.columns = ["PC1", "PC2", "PC3"]
```

```python
from mpl_toolkits.mplot3d import Axes3D
plt.suptitle('Iris plot', fontsize=15)
ax = Axes3D(plt.figure(figsize=(10, 9)), rect=[.01, 0, 0.95, 1], elev=30, azim=134)
ax.scatter(new_data['PC1'], new_data['PC2'], new_data['PC3'], c = targets, s=20)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
```
